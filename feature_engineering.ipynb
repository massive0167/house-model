Missing Values

```python
# let capture as nan values 
# handle categorial data 

feature_nan = [feature for feature in df.columns if df[feature].isnull().sum() > 1 and df[feature].dtype == 'O']

for feature in feature_nan:
    print("{} : {} % missing values.".format(feature, np.round(df[feature].isnull().mean(), 4)))
```

```python
df[feature_nan].isnull().sum()
```

```python
# replacing new value with a new label 
def replace_cat_feature(df, feature_nan):
    data = df.copy()
    data[feature_nan] = data[feature_nan].fillna('Missing')
    return data

df = replace_cat_feature(df, feature_nan)
df[feature_nan].isnull().sum()
```

```python
df.head()
```

```python
## Now lets check for numerical variables the contains missing values
numerical_with_nan = [feature for feature in df.columns if df[feature].isnull().sum() > 1 and df[feature].dtype != 'O']

## We will print the numerical nan variables and percentage of missing values

for feature in numerical_with_nan:
    print("{} : {} % missing".format(feature, np.round(df[feature].isnull().mean(), 4)))
```

```python
# Replacing the numerical missing value '

for feature in numerical_with_nan:
    ## we will replace by median since their outliers 
    median_value = df[feature].median()

    ## create a new feature to capture the value of nan values 
    df[feature + 'nan'] = np.where(df[feature].isnull(), 1, 0)
    df[feature].fillna(median_value, inplace = True)

df[numerical_with_nan].isnull().sum()
```

```python
df
```

```python
## temporal variables 
## remove temopral varibles 

for feature in ['YearBuilt', 'YearRemodAdd', 'GarageYrBlt']:
    df[feature] = df['YrSold'] - df[feature]

df.head()
```

```python
df[['YearBuilt', 'YearRemodAdd', 'GarageYrBlt']].head()
```

```python
df.head()
```

```python
## numerical varibles 
## Since numerical varibles are skewed we will perform log normal distribution 

num_feature = ['LotFrontage', 'LotArea', '1stFlrSF', 'GrLivArea', 'SalePrice']

for feature in num_feature:
    df[feature] = np.log(df[feature])

df.head()
```

```python
categorical_feature = [feature for feature in df.columns if df[feature].dtype == 'O']

categorical_feature
```

```python
## Handling Rare Categorical Feature
## We will remove categorical variables that are present less than 1% of the observations

for feature in categorical_feature:
    temp = df.groupby(feature)['SalePrice'].count()/len(df)
    temp_df = temp[temp > 0.01].index
    df[feature] = np.where(df[feature].isin(temp_df), df[feature], 'Rare_var')
```

```python
df
```

```python
for feature in categorical_feature:
    labels_ordered=df.groupby([feature])['SalePrice'].mean().sort_values().index
    labels_ordered={k:i for i,k in enumerate(labels_ordered,0)}
    df[feature]=df[feature].map(labels_ordered)

df.head(10)
```

```python
feature_scale = [feature for feature in df.columns if feature not in ['Id', 'SalePrice']]


from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
scaler.fit(df[feature_scale])
```

```python
# transform the train and test set, and add on the Id and SalePrice variables

data = pd.concat([df[['Id', 'SalePrice']].reset_index(drop = True), pd.DataFrame(scaler.transform(df[feature_scale]), columns=feature_scale)], axis = 1)
```
